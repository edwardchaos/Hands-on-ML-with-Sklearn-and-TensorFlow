```python
import tensorflow as tf
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris_data = load_iris()
iris_data.keys()
```

```python
X = iris_data['data'][:,(2,3)] # Only care about the last 2 features
y = iris_data['target']
y = (y==0).astype(np.int) #Change to boolean targets
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   random_state=42)
```

```python
feature_cols = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300,100],
                                        n_classes=2,
                                        feature_columns=feature_cols)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf)
dnn_clf.fit(X_train,y_train, batch_size=50,steps=40000)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = dnn_clf.predict(X_test)
y_pred = (y_pred['logits']>0).astype(np.int).reshape(38,)

accuracy_score(y_test,y_pred)
```
